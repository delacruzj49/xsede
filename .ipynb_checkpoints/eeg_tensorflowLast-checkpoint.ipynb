```python
import pandas as pd
import numpy as np
from keras.layers import Dense, Dropout
from keras.models import Sequential
```

```python
df1 = pd.read_csv("eegIDRecord1.csv")
display(df1.head())
df2 = pd.read_csv("eegIDRecord2.csv")
display(df2.head())
df3 = pd.read_csv("eegIDRecord3.csv")
df3.head()
```

```python
OUTPUT_COL = "eegRawValue"
```

```python
df = pd.concat((df1, df2, df3))
df = df.drop(["timestampMs", "location", "tagEvent"], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(OUTPUT_COL, axis=1), df[OUTPUT_COL], test_size=0.25)
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(X_train.shape[1],)))
model.add(Dropout(0.2))

model.add(Dense(125))
model.add(Dropout(0.5))

model.add(Dense(125))
model.add(Dropout(0.5))
          
model.add(Dense(1))
          
model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=256)
```
